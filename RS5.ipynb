```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
booksen=pd.read_excel("C:\\Users\\farno\\Desktop\\booksen.xlsx",index_col=False)
booksen.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
booksen['tags'] = booksen['Genre']+' '+ booksen['About']+' '+ booksen['Author']+' '+ booksen['Genre']+' '+booksen['Genre']
```

```python
vector = cv.fit_transform(booksen["tags"]).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(book):
    index = booksen[booksen['Name'] == book].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:10]:
        print(booksen.iloc[i[0]].Name)
        
```

```python
recommend('سال بلوا')
```
